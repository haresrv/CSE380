```python
# import numpy as np
import pandas as pd
```

```python
#creating dataframe method 1
data = pd.DataFrame({'Country': ['Russia','Colombia','Chile','Equador','Nigeria'],
                    'Rank':[121,40,100,130,11]})
data
```

```python
#creating dataframe method 2
s1= pd.Series(['a', 'a', 'a', 'b','b', 'b', 'c', 'c','c'])
s2=[4, 3, 12, 6, 7.5, 8, 3, 5, 6]
data1 = pd.DataFrame({'group':s1,'ounces':s2})
data1
```

```python
#creating Dataframe method 3
people_dict = { "weight": pd.Series([68, 83, 112],index=["alice", "bob", "charles"]),   "birthyear": pd.Series([1984, 1985, 1992], index=["bob", "alice", "charles"], name="year"),
"children": pd.Series([0, 3], index=["charles", "bob"]),
"hobby": pd.Series(["Biking", "Dancing"], index=["alice", "bob"]),}
people=pd.DataFrame(people_dict)
print (people)
```

```python
#filtering records
people[people["birthyear"] < 1990]
```

```python
#We can do a quick analysis of any data set using:
people.describe()
```

```python
#Among other things, it shows the data set has m rows and n columns with their respective names.
people.info()
```

```python
#create another data with duplicated rows
data3 = pd.DataFrame({'k1':['one']*3 + ['two']*4, 'k2':[3,2,1,3,3,4,4]})
data3

```

```python
#drops repeated rows with same data for the all coulmns
data3.drop_duplicates()
data3
```

```python
data3.drop_duplicates(subset='k1')
```

```python
#adding new column to a dataframe
data3['k3']=[True,False,True,True,False,True,False]
```

```python
#creating dataframe method4
import numpy as np
data4 = pd.DataFrame(np.arange(12).reshape((3, 4)),index=['Ohio', 'Colorado', 'New York'],columns=['one', 'two', 'three', 'four'])
data4
```

```python
#Using rename function
data4.rename(index = {'Ohio':'SanFrancisco'}, columns={'one':'one_p','two':'two_p'},inplace=True)
data4
```

```python
#You can also use string functions
data4.rename(index = str.upper, columns=str.title,inplace=True)
data4
```

```python
#replaing -999 and -1000 with NaN
data5 = pd.Series([1., -999., 2., -999., -1000., 3.])
data5.replace([-999, -1000], np.nan,inplace=True)
data5
```

```python
#slicing dataframe
dates = pd.date_range('20130101',periods=6)
df = pd.DataFrame(np.random.randn(6,4),index=dates,columns=list('ABCD'))
df
```

```python
#get first n rows from the data frame
df[:3]
```

```python
#slice based on date range
df['20130101':'20130104']
```

```python
#slicing based on column names
df.loc[:,:]
```

```python
#slicing based on index of columns
# df.iloc[3] #returns 4th row (index is 3rd)
df.head()
```

```python
#returns a specific range of rows
df.iloc[2:4, 0:2]
```

```python
#returns specific rows and columns using lists containing columns or row indexes
df.iloc[[1,5],[0,2]] 
```

```python
#we can copy the data set
df2 = df.copy()
df2['E']=['one', 'one','two','three','four','three']
df2
```

```python
#select all rows except those with two and four
df2[~df2['E'].isin(['two','four'])]
```

```python
#using OR condition
df.query('A < B | C > A')
```
